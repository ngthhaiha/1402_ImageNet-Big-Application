```python
import os
from pathlib import Path
import shutil
import subprocess
from subprocess import TimeoutExpired

# ================== Cáº¤U HÃNH CHUNG ==================
alphapose_root = Path("/kaggle/working/AlphaPose")

# TRAIN: video input
train_videos_dir = Path("/kaggle/input/cuhk-avenue-dataset/Avenue_Dataset/Avenue Dataset/testing_videos")

# ThÆ° má»¥c output chung
out_root = Path("/kaggle/working/alphapose_out")
train_out_root = out_root / "train"
train_out_root.mkdir(parents=True, exist_ok=True)

# Script / config / checkpoint
cfg_path     = alphapose_root / "configs/coco/resnet/256x192_res152_lr1e-3_1x-duc.yaml"
checkpoint_path = Path("/kaggle/input/pretrained-models/fast_421_res152_256x192.pth")


# ================== HÃM CHáº Y ALPHAPOSE ==================
def run_alphapose_video(video_path: Path, out_json_path: Path, timeout_sec: int = 300):
    """
    Cháº¡y AlphaPose cho 1 video.
    Náº¿u quÃ¡ timeout_sec giÃ¢y mÃ  tiáº¿n trÃ¬nh váº«n chÆ°a xong -> kill vÃ  bá» qua video ÄÃ³.
    """
    vid_name = video_path.stem
    temp_out = Path(f"/kaggle/working/alphapose_temp_video_{vid_name}")
    temp_out.mkdir(parents=True, exist_ok=True)

    print(f"\n=========== Äang cháº¡y VIDEO: {video_path.name} ===========")

    cmd = [
        "python",
        "scripts/demo_inference.py",
        "--cfg", str(cfg_path),
        "--checkpoint", str(checkpoint_path),
        "--outdir", str(temp_out),
        "--video", str(video_path),
        "--sp",
    ]

    try:
        subprocess.run(
            cmd,
            cwd=str(alphapose_root),
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL,
            timeout=timeout_sec,
            check=False,
        )
    except TimeoutExpired:
        print(f"â  TIMEOUT ({timeout_sec}s), bá» qua video {vid_name}")
        shutil.rmtree(temp_out, ignore_errors=True)
        return

    default_json = temp_out / "alphapose-results.json"
    if default_json.exists():
        shutil.move(str(default_json), str(out_json_path))
        print(f"â ÄÃ£ lÆ°u: {out_json_path}")
    else:
        print(f"â  KhÃ´ng tÃ¬m tháº¥y alphapose-results.json cho video {vid_name}")

    shutil.rmtree(temp_out, ignore_errors=True)


# ================== CHáº Y CHO TRAIN (Tá»ª 09_008 TRá» ÄI) ==================
train_videos = sorted(
    [v for v in train_videos_dir.iterdir() if v.suffix.lower() in [".mp4", ".avi", ".mov"]]
)

print("\n===== TRAIN VIDEOS =====")
print("Tá»ng sá» video train:", len(train_videos))

# tÃ¬m index báº¯t Äáº§u tá»« video 09_008
start_name = "01"
start_idx = 0
for i, v in enumerate(train_videos):
    if v.stem == start_name:
        start_idx = i
        break
else:
    print(f"KhÃ´ng tÃ¬m tháº¥y video {start_name} trong thÆ° má»¥c train.")
    start_idx = len(train_videos)  # khÃ´ng cháº¡y gÃ¬

print(f"Báº¯t Äáº§u cháº¡y tá»« video cÃ³ index {start_idx}: {start_name} (náº¿u tá»n táº¡i)\n")

for video_path in train_videos[start_idx:]:
    vid_name = video_path.stem
    out_json = train_out_root / f"{vid_name}_alphapose_results.json"
    run_alphapose_video(video_path, out_json, timeout_sec=300)

print("\n=== HoÃ n táº¥t AlphaPose cho TRAIN (tá»« 01 trá» Äi) ===")
print("JSON train náº±m trong:", train_out_root)
```
